La nafta súper es la más comun y por eso vamos a filtrar y analizar éste producto primero.

```python
boolsuper = (surtidoresH.loc[:,'producto'] == 'Nafta (súper) entre 92 y 95 Ron')#bool: es nafta super?
superind = boolsuper[boolsuper == True].index #array de indices de producto super
nafsuper = surtidoresH.iloc[superind]#dataframe sólo nafta súper

nafsuper.fecha_vigencia
```

```python

nafsuperordentemp = np.array(nafsuper.fecha_vigencia.sort_values().index)
timeordered = pd.DataFrame(nafsuper, index=nafsuperordentemp)
promediotot = np.average(np.array(timeordered.precio))
timeordered.index = np.arange(252492)
timeordered
x=timeordered['fecha_vigencia']
y=timeordered['precio']
timeordered
timeordered[timeordered['idempresa']==7566]

#recordar: loc=ubicación explícita utilizada al definir el DataFrame o la Serie. Es decir, en la primera parte va
#[nuevamente entre corchetes] el índice correspondiente al valor en cuestion que fue definido explicitamente por 
#el dataframe o la serie. Luego de la coma va entre corchetes también la clave asignada en su diccionario análogo.
#          sin embargo, iloc va con los indices implicitos clasicos que son asignados a todos los objetos iterables
#de python. en la primera parte va el indice implicito de la fila y en la segunda el indice implucito asignado a la clave
#del diccionario analogo.


```
